```python
import tensorflow as tf
from keras import Model
from keras.layers import Conv2D, Conv2DTranspose, BatchNormalization, MaxPooling2D, Input
from keras.datasets import mnist
from keras import backend as K

import matplotlib.pyplot as plt
import numpy as np
```

```python
def dice_coef(y_true, y_pred):
    return (2. * K.sum(y_true * y_pred) + 1.) / (K.sum(y_true) + K.sum(y_pred) + 1.)
```

```python
def base_autoencoder(level = 2, input_size=(28, 28, 1), kernel_list=[32, 64]):
    img_input = Input(input_size)
    
    i = 0
    while True:
        x = Conv2D(kernel_list[i], (3, 3), padding='same', activation='relu')(img_input if i == 0 else x)
        x = BatchNormalization()(x)
        x = Conv2D(kernel_list[i], (3, 3), padding='same', activation='relu')(x)
        x = BatchNormalization()(x)
        x = MaxPooling2D()(x)
        i += 1
        if i == level:
            break
            
    bn = x
    print(bn)
    
    i = level-1
    while True:
        x = Conv2DTranspose(kernel_list[i], (2, 2), strides=(2, 2), padding='same', activation='relu')(x)
        x = BatchNormalization()(x)
        x = Conv2D(kernel_list[i], (3, 3), padding='same', activation='relu')(x)
        x = BatchNormalization()(x)
        x = Conv2D(kernel_list[i], (3, 3), padding='same', activation='relu')(x)
        x = BatchNormalization()(x)
        i -= 1
        if i < 0:
            break
            
    x = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)
    
    model = Model(img_input, x)
    bn_model = Model(img_input, bn)
    
    model.compile(optimizer=tf.optimizers.Adam(), loss=tf.losses.mean_squared_error, metrics=[dice_coef])
    
    return model, bn_model
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train = x_train.astype('float32')/255
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
```

```python
noizes = [0.1, 0.4, 0.9]

# зашумлённые данные
ndata = []
for noize in noizes:
    _y_train = x_train.copy()
    _x_train = x_train.copy()
    _x_train += np.random.randn(_x_train.shape[0], _x_train.shape[1], _x_train.shape[2], _x_train.shape[3]) * noize
    ndata.append((_x_train, _y_train, noize))
```

```python
# смотрим на наш шум
for d in ndata:
    print(f'Уровень шума %s' % (d[2]))
    plt.imshow(d[0][4,:,:,0])
    plt.show()
```

```python
print('Исходный образ')
plt.imshow(d[1][4,:,:,0])
plt.show()
```

```python
# параметры
epochs = 10
batch_size = 100
validation_split = 0.2
```

```python
# обучаем модели
models = []
for d in ndata:
    model, model_bn = base_autoencoder(level = 2, input_size=(28,28,1), kernel_list=[32,64])
    hist = model.fit(d[0][:10000], d[1][:10000], epochs=epochs, batch_size=batch_size, validation_split=validation_split)
    
    models.append((model, model_bn, hist))
```

```python
# графики обучения моделей по уровню шума от низкого к высокому
for mi, m in enumerate(models):
    print(f'Уровень шума %s' % (ndata[mi][2]))
    plt.plot(m[2].history['val_dice_coef'],'r',label='val_acc')
    plt.plot(m[2].history['dice_coef'],'g',label='train_acc')
    plt.legend()
    plt.show()
```

```python
# как предсказания?
for mi, m in enumerate(models):
    _model, _nb_model, hist = m
    _x_train, _y_train, _noize = ndata[mi]
    
    print(f'Уровень шума %s' % _noize)
    mask = np.where(y_train == 0)
    y_pred = _model.predict(_x_train[mask[0]])
    
    ff = _nb_model.predict(x_train[mask[0]])
    # print(ff.shape)
    
    ff = np.mean(ff, axis=0)
    # print(ff.shape)
    
    print('Входные данные')
    plt.imshow(_x_train[mask[0]][1,:,:,0])
    plt.show()
    print('Что мы получили')
    plt.imshow(y_pred[1,:,:,0])
    plt.show()
    print('Что мы ожидали')
    plt.imshow(_y_train[mask[0]][1,:,:,0])
    plt.show()
    
    print('А что в "бутылочном горлышке"?')
    g1=ff[:,:,0]
    plt.subplot(311)
    plt.imshow(g1)
    g1=ff[:,:,1]
    plt.subplot(312)
    plt.imshow(g1)
    g1=ff[:,:,2]
    plt.subplot(313)
    plt.imshow(g1)
    plt.show()
```

MNIST - очень хороший датасет и даже сильно зашумлённые данные прекрасно распознаются.

```python
# возьмём модель, обученную на низком уровне шума и дадим ей на вход сильно зашумлённые данные
m_indeces = [0, 1] # модели, обученные на 0.1 и 0.4 шума
d_indeces = [1, 2] # данные, с 0.4 и 0.9 шума

for mi in m_indeces:
    _model, _nb_model, hist = models[mi]
    print(f'Уровень шума модели %s' % ndata[mi][2])
    
    for di in d_indeces:
        # пропускаем одинаковый шум
        if ndata[mi][2] == ndata[di][2]:
            continue
        
        _x_train, _y_train, _noize = ndata[di]

        print(f'Уровень шума данных %s' % _noize)
        mask = np.where(y_train == 0)
        y_pred = _model.predict(_x_train[mask[0]])

        print('Входные данные')
        plt.imshow(_x_train[mask[0]][1,:,:,0])
        plt.show()
        print('Что мы получили')
        plt.imshow(y_pred[1,:,:,0])
        plt.show()
        print('Что мы ожидали')
        plt.imshow(_y_train[mask[0]][1,:,:,0])
        plt.show()
```

Как мы видим, модели обученные на низком уровне зашумления по-прежнему неплохо справляются с шумом средней силы  
Аналогичная ситуация и с моделями для среднего шума - они неплохо понимают сильно защумлённые данные
